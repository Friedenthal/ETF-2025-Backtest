# Scores Generation File

1. Generates the Old Regression score summary & saves the file into ./Score Summary/df_score_summary_old_reg.csv
2. Generates the New Regression score summary & saves the file into ./Score Summary/df_score_summary_new_reg.csv

```python
# Install the necessary packages from "_packages_startup.txt"

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import datetime as dt
import itertools as it
import ast

import sys
sys.path.append('../src')
from utils import *
```

```python
# Import the price data from "../Raw Data/ETF_PriceFile_2017_to_2025.csv"
df_price_all = pd.read_csv('../Raw Data/ETF_PriceFile_2017_to_2025.csv', index_col=0)
```

```python
# Dates are in Excel Date format. Arrange them in descending order to send generate scores. Then view the input to confirm!
df_price_all.sort_index(ascending=False, inplace=True)
df_price_all.head(5)
```

```python
# Generate the Old Regression Summary dataframe. The regression_array list can be found under documentation in  ../src/utils.py 

regression_array_old = list(range(10, 126, 5))
df_score_summary_old_reg = Regression_Gen(df_price_all, regression_array_old)
```

```python
# Generate the New Regression Summary dataframe. The regression_array list can be found under documentation in  ../src/utils.py

regression_array_current = [25, 30, 40, 55, 75, 100, 130, 165, 205, 250]
df_score_summary_new_reg = Regression_Gen(df_price_all, regression_array_current)
```

```python
# Set the index to 'Date' for both dataframes
df_score_summary_old_reg.set_index('Date', inplace=True)
df_score_summary_new_reg.set_index('Date', inplace=True)

# Export the Score Summary dataframes to CSV files in "../Score Summary/df_score_summary_old_reg.csv" and "../Score Summary/df_score_summary_new_reg.csv"
df_score_summary_old_reg.to_csv('../Score Summary/df_score_summary_old_reg.csv')
df_score_summary_new_reg.to_csv('../Score Summary/df_score_summary_new_reg.csv')
```
